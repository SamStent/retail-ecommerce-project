> ## _Notebook para análisis_

```python
import pandas as pd
from sqlalchemy import create_engine
import plotly.express as px
import plotly.graph_objects as go

print("Entorno funcionando correctamente")

# Conexión a PostgreSQL
engine = create_engine("postgresql://postgres:postgres@localhost:5432/retail")

# Carga de vistas principales
sales_summary = pd.read_sql("SELECT * FROM vw_sales_summary", engine)
monthly_sales = pd.read_sql("SELECT * FROM vw_monthly_sales", engine)
product_perf = pd.read_sql("SELECT * FROM vw_product_performance", engine)
city_sales = pd.read_sql("SELECT * FROM vw_city_sales", engine)
customer_value = pd.read_sql("SELECT * FROM vw_customer_value", engine)
category_sales = pd.read_sql("SELECT * FROM vw_category_sales", engine)
```

> # _KPIs ejecutivas_

```python
# KPI 1 — Revenue total
total_revenue = sales_summary['order_revenue'].sum()
total_revenue
```

```python
# KPI 2 — Número total de órdenes
total_orders = sales_summary['order_id'].nunique()
total_orders
```

```python
# KPI 3 — Ticket promedio
avg_ticket = sales_summary['order_revenue'].mean()
round(avg_ticket, 2)
```

```python
# KPI 4 — Valor promedio por cliente
avg_customer_value = customer_value['total_spent'].mean()
round(avg_customer_value, 2)
```

```python
# KPI 5 — Ciudad con mayor revenue
city_sales.sort_values("revenue", ascending=False).head(1)
```

```python
# KPI 6 — Rating promedio general
avg_rating = product_perf['avg_rating'].mean()
round(avg_rating)
```

> ### _Gráficos con Plotly_

```python
# Top 10 productos por revenue
top_products = product_perf.sort_values("revenue", ascending=False).head()
top_products
```

```python
# Gráfico
fig = px.bar(
    top_products,
    x="revenue",
    y="product_name",
    orientation="h",
    title="Top 10 productos por revenue"
)
fig.show()
```

```python
# Revenue por categoría
category_sales.sort_values("revenue", ascending=False)
```

```python
# Gráfico
fig = px.bar(
    category_sales.sort_values("revenue", ascending=False),
    x="product_category",
    y="revenue",
    title="Revenue por categoría"
)
fig.show()
```

```python
# Distribución del gasto por cliente gráfico
fig = px.box(
    customer_value,
    y="total_spent",
    title="Distribución del gasto total por cliente"
)
fig.show()
```

```python
# Revenue por ciudad
fig = px.bar(
    city_sales.sort_values("revenue", ascending=False),
    x="city",
    y="revenue",
    title="Revenue por ciudad"
)
fig.show()
```

> ## _🟦 Insights ejecutivos_

El revenue total del dataset es X, con un ticket promedio de Y.

La ciudad con mayor revenue es A, lo que sugiere una concentración geográfica relevante.

Los productos más rentables pertenecen a la categoría B, indicando una preferencia clara del mercado.

El revenue mensual muestra una tendencia C, lo que podría relacionarse con estacionalidad.

El gasto por cliente es altamente desigual, con un pequeño grupo de clientes generando la mayor parte del revenue.
